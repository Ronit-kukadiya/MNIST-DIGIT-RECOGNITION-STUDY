# MNIST Dataset - Exploratory Data Analysis (EDA)


### Introduction

### This notebook performs Exploratory Data Analysis (EDA) on the MNIST dataset. We will explore the dataset structure, visualize some sample images, and analyze pixel distributions.


```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
train_df = pd.read_csv("../0DATA_SELECTION/CSV_data/train_data.csv")
test_df = pd.read_csv("../0DATA_SELECTION/CSV_data/test_data.csv")

# display first few rows
train_df.head(10)
```

```python
# Shape of the datasets
print("Train Data Shape:", train_df.shape)
print("Test Data Shape:", test_df.shape)

# Check for missing values
print("Missing values in train data:", train_df.isnull().sum().sum())
print("Missing values in test data:", test_df.isnull().sum().sum())
```

```python
# Count of each digit in the dataset
plt.figure(figsize=(8, 4))
sns.countplot(x=train_df['label'], palette="viridis")
plt.title("Distribution of Digits in Training Data")
plt.xlabel("Digit Label")
plt.ylabel("Count")
plt.show()
```

```python
# Summary statistics of pixel values
pixel_stats = train_df.iloc[:, 1:].describe()
pixel_stats
```

```python
# Flatten pixel values and plot histogram
# this shows which color value is found more in datasets (from 0 to 255)
pixel_values = train_df.iloc[:, 1:].values.flatten()

plt.figure(figsize=(8, 4))
plt.hist(pixel_values, bins=50, color='green', alpha=0.7)
plt.title("Pixel Intensity Distribution (0-255)")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Function to visualize random train images
def show_images(df, num_images=6):
    plt.figure(figsize=(10, 4))
    for i in range(num_images):
        img = df.iloc[i, 1:].values.reshape(28, 28)
        label = df.iloc[i, 0]
        plt.subplot(2, 3, i+1)
        plt.imshow(img, cmap="gray")
        plt.title(f"Label: {label}")
        plt.axis("off")
    plt.show()

# Display sample images
show_images(train_df)
```

```python
# Function to visualize random test images
def show_images(df, num_images=6):
    plt.figure(figsize=(10, 4))
    for i in range(num_images):
        img = df.iloc[i, 1:].values.reshape(28, 28)
        label = df.iloc[i, 0]
        plt.subplot(2, 3, i+1)
        plt.imshow(img, cmap="gray")
        plt.title(f"Label: {label}")
        plt.axis("off")
    plt.show()

# Display sample images
show_images(test_df)
```

```python
# Compute mean image
mean_image = train_df.iloc[:, 1:].mean().values.reshape(28, 28)

plt.figure(figsize=(6, 6))
sns.heatmap(mean_image, cmap="coolwarm", annot=False)
plt.title("Mean Image Heatmap")
plt.show()
```

```python
# Compute median image
median_image = train_df.iloc[:, 1:].median().values.reshape(28, 28)

plt.figure(figsize=(6, 6))
sns.heatmap(median_image, cmap="coolwarm", annot=False)
plt.title("Mean Image Heatmap")
plt.show()
```

```python
# Compute min image
median_image = train_df.iloc[:, 1:].min().values.reshape(28, 28)

plt.figure(figsize=(6, 6))
sns.heatmap(median_image, cmap="coolwarm", annot=False)
plt.title("Mean Image Heatmap")
plt.show()
```

```python
# Compute max image
median_image = train_df.iloc[:, 1:].max().values.reshape(28, 28)

plt.figure(figsize=(6, 6))
sns.heatmap(median_image, cmap="coolwarm", annot=False)
plt.title("Mean Image Heatmap")
plt.show()
```

```python
print("Min pixel value:", train_df.iloc[:, 1:].min().min())
print("Max pixel value:", train_df.iloc[:, 1:].max().max())
print("Median pixel value:", train_df.iloc[:, 1:].median().median())
print("Mean pixel value:", train_df.iloc[:, 1:].mean().mean())
```
